```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("student_data.csv")
```

```python
df.head()
```

```python
plt.plot(df['sex'],df['age'])
```

```python
plt.xlabel("sex")

plt.ylabel("age")

plt.title("Age and Sex")

plt.plot(df['age'],df['sex'])
plt.xlabel("sex")
plt.ylabel("age")
plt.title("Age And Sex")
plt.show()
```

```python
df['sex'].value_counts().plot(kind='bar')
plt.title("Sex in bar")
plt.show()
```

```python
plt.hist(df['age'], bins=5, color = 'skyblue')
plt.title("Age and Sex")
plt.xlabel('age')
plt.ylabel('count')
plt.show()
```

```python
df['address'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Address in pie")
plt.show()
```

```python
plt.scatter(df['age'], df['sex'], color='green')
plt.xlabel("Age")
plt.ylabel("Sex")
plt.title("Age And Sex")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.hist(df['age'])
plt.title("Age")

plt.subplot(1,2,2)
plt.hist(df['sex'])
plt.title("Sex")

plt.show()
```

```python
df[df['sex']=="M"]['age'].plot(kind='hist', alpha=0.5, label='Male')
df[df['sex']=="F"]['age'].plot(kind='hist', alpha=0.5, label='Female')
plt.legend()
plt.title("Grades Distribution by Gender")
plt.show()
```

```python
df['Pstatus'].value_counts().plot(kind='bar')
plt.show()
```
